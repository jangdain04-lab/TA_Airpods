```python
import pandas as pd

knu_path = "/content/KnuSentiLex/SentiWord_Dict.txt"

# 탭(\t)으로 구분된 텍스트 파일이라 sep="\t"
knu_df = pd.read_csv(knu_path, sep="\t", header=None)

print(knu_df.head())
print(knu_df.shape)
```

```python
knu_df.columns = ["word", "polarity"]
knu_df.head()
```

```python
knu_dict = dict(zip(knu_df['word'], knu_df['polarity']))
```

```python
def get_sentiment_score(text):
    score = 0
    for word, pol in knu_dict.items():
        if word in text:
            score += pol
    return score
```

```python
df_clean["sentiment"] = df_clean["lyric"].apply(get_sentiment_score)
df_clean.head()
```

```python
import pandas as pd

url_wc = "https://raw.githubusercontent.com/yeon524/kpop-lyrics-analytics/main/datasets/lyrics_word_count_konlpy.csv"
wc = pd.read_csv(url_wc)
wc.head()
```

```python
content_pos = ["Noun", "Adj", "Verb"]
wc_content = wc[wc["type"].isin(content_pos)]
```

```python
wc_top = wc_content.sort_values("count", ascending=False).head(3000)
```

```python
# 1) KnuSentiLex 로드
knu_path = "/content/KnuSentiLex/SentiWord_Dict.txt"
knu = pd.read_csv(knu_path, sep="\t", header=None, names=["word", "polarity", "score"])

# 2) 우리 corpus 상위 단어와 교집합
cand = wc_top.merge(knu, left_on="word", right_on="word", how="inner")
```

```python
print(cand.shape)
cand.head(20)
```

```python
# 점수 절댓값 컬럼 추가
cand["abs_score"] = cand["score"].abs()

# 예시: 감정 점수 절댓값 >= 1 이고, 코퍼스에서 최소 50번 이상 등장한 단어만
candidates = cand[(cand["abs_score"] >= 1.0) & (cand["count"] >= 50)]

print(candidates.shape)
candidates.sort_values("score", ascending=False).head(30)   # 높은 긍정
candidates.sort_values("score").head(30)                    # 높은 부정
```

```python
candidates = cand[cand["count"] >= 30]
```

```python
candidates.shape
candidates.head()
```

```python
# 감정 후보 단어만 필요한 컬럼만 추려서 저장
cols = ["word", "type", "count", "polarity"]
candidates[cols].to_csv("emotion_candidates_v1.csv", index=False)
```
