```python
import re
from kiwipiepy import Kiwi
kiwi = Kiwi()

# ==========================================================
# 1) 문장 분리
# ==========================================================
def lyrics_to_corpus(lyrics):
    if not isinstance(lyrics, str):
        return []
    corpus = re.split(r'\r\n|\n', lyrics)
    corpus = [s.strip() for s in corpus if s.strip() != ""]
    return corpus


# ==========================================================
# 2) 단순 Clean
# ==========================================================
def clean_sentence(sentence):
    sentence = re.sub(r'[^가-힣A-Za-z0-9\s]', ' ', sentence)
    sentence = re.sub(r'\s+', ' ', sentence)
    return sentence.strip()


# ==========================================================
# 3) 단어 단위 노이즈 제거
# ==========================================================
noise_words = {
    "oh","yeah","woo","la","na","ah",
    "나나","나나나","나나나나","담다디",
    "오예","아예","아하",
    "say","one","two","three","four"
}

def remove_noise_words(sentence):
    words = sentence.split()
    cleaned = [w for w in words if w.lower() not in noise_words]
    return " ".join(cleaned).strip()


# ==========================================================
# 4) 반복구(후렴) 패턴 감지
# ==========================================================
repeated_noise = [
    "담다디","담다디다담","라라라","나나나","나나","다다",
]

def is_repeated_chorus(sentence):
    words = sentence.split()
    if len(words) == 0:
        return False
    hits = sum(1 for w in words if w in repeated_noise)
    return (hits / len(words)) >= 0.7


# ==========================================================
# 5) 한국어 포함 여부
# ==========================================================
def has_korean(sentence):
    return any("가" <= ch <= "힣" for ch in sentence)


# ==========================================================
# 6) 의미 있는 품사 포함 여부
# ==========================================================
meaningful_pos = {"NNG","NNP","VV","VA"}

def is_meaningful(sentence):
    tokens = kiwi.tokenize(sentence)
    return any(tok.tag in meaningful_pos for tok in tokens)


# ==========================================================
# 7) 최종 문장 전처리
# ==========================================================
def preprocess_sentences(sentences):
    final = []
    for s in sentences:
        s = clean_sentence(s)
        if s == "":
            continue

        s = remove_noise_words(s)
        if s == "":
            continue

        if is_noise(s):
            continue

        if is_repeated_chorus(s):
            continue

        if not has_korean(s):
            continue

        if not is_meaningful(s):
            continue

        final.append(s)

    # 🔥 중복 문장 제거 (순서 유지)
    seen = set()
    unique_final = []
    for s in final:
        if s not in seen:
            unique_final.append(s)
            seen.add(s)

    return unique_final



# ==========================================================
# 8) Kiwi 기반 토큰화 (명사/동사/형용사/부사/접속부사)
# ==========================================================
def tokenize(sentence):
    allowed_tags = {"NNG", "NNP", "VV", "VA", "MAG", "MAJ"}  # 저장할 품사들
    tokens = []

    for tok in kiwi.tokenize(sentence):
        tag = tok.tag

        # 명사・부사는 그대로
        if tag in {"NNG", "NNP", "MAG", "MAJ"}:
            tokens.append(tok.form)

        # 동사・형용사는 기본형(lemma)로 통일
        elif tag in {"VV", "VA"}:
            tokens.append(tok.form + "다")

    # 중복 제거
    tokens = sorted(list(set(tokens)))
    return tokens

```

lyrics → 문장 분리 → 문장 정제 → noise 제거 → 품사 기반 필터 → tokenize()

```python
import random

def debug_show_song(df, n=10):
    sample_idx = random.sample(range(len(df)), n)

    for idx in sample_idx:
        row = df.iloc[idx]
        title = row['title']
        singer = row['singer']
        lyric = row['lyric']

        print("="*120)
        print(f"🎵 제목: {title}   |   🎤 아티스트: {singer}")
        print("-"*120)

        # 1) 문장 분리
        sentences = lyrics_to_corpus(lyric)

        print("\n[원본 문장들]")
        for s in sentences:
            print(" ·", s)

        # 2) 전처리 적용
        processed = preprocess_sentences(sentences)

        print("\n[전처리 후 문장들]")
        for s in processed:
            print(" →", s)

        # 3) 토큰화
        print("\n[토큰화 결과]")
        for s in processed:
            tokens = tokenize(s)
            print("   🔸", tokens)

        print("\n")

# 실행
debug_show_song(df_kpop, n=10)
```
