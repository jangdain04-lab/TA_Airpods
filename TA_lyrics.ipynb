```python
import pandas as pd
import numpy as np
import re

# ----------------------------
# 1. 한국어 감정 사전 로드
# ----------------------------
lexicon_path = "/content/lyrics-emotion-analysis/korean/API/asset/vocab_9class_500.csv"

# cp949 인코딩으로 읽기
lexicon = pd.read_csv(lexicon_path, encoding="cp949")

# 단어 컬럼 이름 정리
lexicon = lexicon.rename(columns={"Unnamed: 0": "word"})

# 감정 점수 컬럼 목록 (word 빼고 나머지 전부)
emotion_cols = [c for c in lexicon.columns if c != "word"]

print("감정 사전 크기:", lexicon.shape)
print("감정 점수 컬럼:", emotion_cols)

lexicon.head()
```

```python
# ----------------------------
# 2. K-pop 가사 데이터 로드
# ----------------------------
df_kpop = pd.read_csv(
    "/content/kpop-lyrics-analytics/datasets/lyrics_by_year_1964_2023.csv"
)

print(df_kpop.shape)
df_kpop.head()
```

```python
# ----------------------------
# 3. 가사 텍스트 전처리
#    - NaN -> 빈 문자열
#    - 한글/공백만 남기기
# ----------------------------

def clean_lyrics(text):
    if pd.isna(text):
        return ""
    # 한글 + 공백만 남기기
    text = re.sub(r"[^가-힣\s]", " ", str(text))
    # 여러 칸 공백 -> 한 칸
    text = re.sub(r"\s+", " ", text).strip()
    return text

df_kpop["lyrics_clean"] = df_kpop["lyric"].apply(clean_lyrics)

df_kpop[["year", "title", "singer", "lyrics_clean"]].head()
```

```python
# ----------------------------
# 4. word -> {감정1: 점수, 감정2: 점수, ...} 형태 딕셔너리
# ----------------------------
lexicon_dict = (
    lexicon.set_index("word")[emotion_cols]
           .to_dict(orient="index")
)

# 구조 확인
list(lexicon_dict.items())[:5]
```

```python
# ----------------------------
# 5. 한 곡(한 줄 가사)에 대해 감정 점수 계산 함수
# ----------------------------
def score_lyrics_with_lexicon(text):
    """
    text: 전처리된 가사 문자열 (lyrics_clean)
    반환: 각 감정별 sum, mean + 매칭된 단어 개수(lexicon_hits)
    """
    if not isinstance(text, str) or text.strip() == "":
        # 가사가 없으면 모두 0
        result = {f"{col}_sum": 0.0 for col in emotion_cols}
        result.update({f"{col}_mean": 0.0 for col in emotion_cols})
        result["lexicon_hits"] = 0
        return pd.Series(result)

    tokens = text.split()

    # 합 계수 초기화
    scores_sum = {col: 0.0 for col in emotion_cols}
    hit_count = 0

    for tok in tokens:
        if tok in lexicon_dict:
            hit_count += 1
            for col in emotion_cols:
                scores_sum[col] += lexicon_dict[tok][col]

    # 평균 점수 계산
    result = {}
    for col in emotion_cols:
        result[f"{col}_sum"] = scores_sum[col]
        if hit_count > 0:
            result[f"{col}_mean"] = scores_sum[col] / hit_count
        else:
            result[f"{col}_mean"] = 0.0

    result["lexicon_hits"] = hit_count
    return pd.Series(result)
```

```python
# ----------------------------
# 6. 모든 곡에 감정 점수 적용
# ----------------------------
lex_scores = df_kpop["lyrics_clean"].apply(score_lyrics_with_lexicon)

# 원래 df와 합치기
df_lex = pd.concat([df_kpop, lex_scores], axis=1)

df_lex.head()
```

```python
# 감정 사전 단어가 하나라도 나온 곡만
df_lex_nonzero = df_lex[df_lex["lexicon_hits"] > 0].copy()

print("전체 곡 수:", len(df_lex))
print("사전 단어가 나온 곡 수:", len(df_lex_nonzero))
print("비율:", len(df_lex_nonzero) / len(df_lex))
```

```python
# decade 컬럼 만들기 (1960, 1970, 1980, …)
df_lex_nonzero["decade"] = (df_lex_nonzero["year"] // 10) * 10

# mean 컬럼 이름들
mean_cols = [f"{c}_mean" for c in emotion_cols]

decade_emotion = (
    df_lex_nonzero
    .groupby("decade")[mean_cols]
    .mean()
    .reset_index()
)

decade_emotion.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
for col in mean_cols:
    plt.plot(decade_emotion["decade"], decade_emotion[col], marker="o", label=col)

plt.xlabel("Decade")
plt.ylabel("Average emotion score")
plt.title("Emotion trends by decade")
plt.legend()
plt.show()
```

```python
from collections import Counter

# decade별로 단어 빈도수 계산
df_tokens = (
    df_lex_nonzero
    .assign(tokens = df_lex_nonzero["lyrics_clean"].str.split())
    .explode("tokens")
)

# 너무 짧은 단어 제거 (필요하면)
df_tokens = df_tokens[df_tokens["tokens"].str.len() > 1]

def top_words_by_decade(decade, n=20):
    sub = df_tokens[df_tokens["decade"] == decade]
    counter = Counter(sub["tokens"])
    return counter.most_common(n)

# 예시: 1990년대, 2010년대 상위 20개 키워드
print("1990s:", top_words_by_decade(1990, 20))
print("2010s:", top_words_by_decade(2010, 20))
```

```python
# ============================================================
# ① 전처리 함수 집합 (프로젝트에 직접 삽입)
# ============================================================

import re

# 가사 문장별 분리
def lyrics_to_corpus(lyrics):
    lyrics = re.split(r'\r\n|\n', lyrics)
    corpus = list(set(lyrics))
    return corpus

# 문장별 전처리
def sentence_preprocessing(corpus):
    corpus = [re.sub(r'[^\w]', " ", sentence) for sentence in corpus]
    corpus = [sentence.strip() for sentence in corpus]
    corpus = [sentence for sentence in corpus if len(sentence) > 5]
    corpus = [list(set(sentence.split())) for sentence in corpus]
    corpus = [" ".join(sentence) for sentence in corpus]
    return corpus

# 형태소 정규화(Komoran 사용)
def lexicon_tokenizer_komoran(sentence, kmoran):
    temp = []
    for word, tag in kmoran.pos(sentence):
        if tag in ['NNG', 'NNP']:
            temp.append(word)
        elif tag in ['VV', 'VA']:
            temp.append(word + "다")
        elif tag == 'SL':
            temp.append(word.lower())
    return temp

# 한국어/영어 판별
def EnglishOrKorean(input_s):
    k_count = 0
    e_count = 0
    for c in input_s:
        if '가' <= c <= '힣':
            k_count += 1
        elif 'a' <= c.lower() <= 'z':
            e_count += 1
    return "k" if k_count > 1 else "e"

# 한국어 비율 기준 필터
def isKorean_percent(input_s, percentage):
    k_count = 0
    e_count = 0
    for c in input_s:
        if '가' <= c <= '힣':
            k_count += 1
        elif 'a' <= c.lower() <= 'z':
            e_count += 1

    if k_count == 0:
        return 0

    percent = k_count / (k_count + e_count)
    return 1 if percent > percentage else 0

# Instrumental / MR 제거 여부 판단
def isRemoveData(title):
    if 'inst.' in title.lower():
        return 1
    if 'instrumental' in title.lower():
        return 1
    if 'mr' in title:
        return 1
    return 0
```
