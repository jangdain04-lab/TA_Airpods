```python
from konlpy.tag import Okt
okt = Okt()

def tokenize(text):
    return okt.morphs(text)
```

```python
import re

def clean_text(text):
    text = re.sub(r"[^가-힣A-Za-z0-9\s]", " ", str(text))  # 특수문자 제거
    text = re.sub(r"\s+", " ", text)  # 중복 공백 제거
    return text.strip()
```

```python
df_clean['lyric_clean'] = df_clean['lyric'].apply(clean_text)
```

```python
from konlpy.tag import Okt
okt = Okt()

def tokenize(text):
    return okt.morphs(text)
```

```python
df_clean['tokens'] = df_clean['lyric_clean'].apply(tokenize)
```

```python
df_clean['pos'] = df_clean['lyric_clean'].apply(lambda x: okt.pos(x))
```

```python
basic_stop = [
    "그리고", "그러나", "하지만", "또", "또한", "근데",
    "은", "는", "이", "가", "을", "를", "에", "에서", "로", "으로",
    "도", "만", "보다", "처럼", "까지", "에게", "한테", "때문에",
]
```

```python
fillers = [
    "오", "아", "예", "워", "우", "야", "헤이",
    "oh", "yeah", "ooh", "uh", "woo", "la", "lala", "hey",
]
```

```python
weak_verbs = [
    "하다", "하고", "해", "했어", "된다", "되다", "있다", "없다",
    "가다", "오다", "된다", "하게", "되어", "하네", "했네",
]
```

```python
stopwords = list(set(basic_stop + fillers + weak_verbs))
#최종 불용어 리스트
```

```python
df_clean['tokens'] = df_clean['tokens'].apply(
    lambda token_list: [t for t in token_list if t not in stopwords]
)
```

```python
df_clean[['title', 'tokens']].head(10)
```

```python
from collections import Counter

all_tokens = sum(df_clean['tokens'], [])  # 전체 토큰 합치기
Counter(all_tokens).most_common(30)
```

```python
df_clean[['lyric', 'lyric_clean', 'tokens']].head(5)
```

```python
# 1) 불용어 제거까지 완료된 상태에서 실행
def unique_tokens(tokens):
    return list(dict.fromkeys(tokens))  # 순서 유지 + 중복 제거

df_clean['tokens'] = df_clean['tokens'].apply(unique_tokens)

# 2) TF-IDF 등 분석용 문자열 갱신
df_clean['token_str'] = df_clean['tokens'].apply(lambda x: " ".join(x))
```

```python
# 중복 제거 전 원래 tokens 복원
df_clean['tokens_before'] = df_clean['lyric_clean'].apply(tokenize)

# 불용어 제거 다시 적용
df_clean['tokens_before'] = df_clean['tokens_before'].apply(
    lambda token_list: [t for t in token_list if t not in stopwords]
)
```

```python
def unique_tokens(tokens):
    return list(dict.fromkeys(tokens))

df_clean['tokens'] = df_clean['tokens_before'].apply(unique_tokens)
```

```python
df_clean[['title', 'tokens_before', 'tokens']].sample(5)
```

```python
from collections import Counter

top_words_by_year = {}

years = sorted(df_clean['year'].unique())

for y in years:
    tokens = sum(df_clean[df_clean['year'] == y]['tokens'], [])  # 리스트 합치기
    freq = Counter(tokens).most_common(30)
    top_words_by_year[y] = freq

top_words_by_year
```
