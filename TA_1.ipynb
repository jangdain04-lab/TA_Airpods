```python
import pandas as pd
import matplotlib.pyplot as plt

# 깃허브 RAW CSV
url = "https://raw.githubusercontent.com/yeon524/kpop-lyrics-analytics/main/datasets/lyrics_by_year_1964_2023.csv"

df = pd.read_csv(url)

print("원본 데이터:", len(df))
df.head()
```

```python
df = df.dropna(subset=['lyric'])
print("가사 있는 노래 개수 =", len(df))
```

```python
df = df[(df['year'] >= 1970) & (df['year'] <= 2020)]
print("1970~2020 노래 개수 =", len(df))
```

```python
df['decade'] = (df['year'] // 10) * 10
df['decade'] = df['decade'].astype(str) + "s"

df[['year', 'decade']].head()
```

```python
emotion_lexicon = {
    "joy": [
        "행복", "기쁨", "기쁘", "즐겁", "설레", "웃음", "미소",
        "환희", "축하", "환한 미소", "축복", "사랑스러"
    ],
    "sadness": [
        "슬픔", "슬프", "눈물", "울음", "아픔", "괴로움", "그리움",
        "쓸쓸", "외롭", "허전", "상처", "비통", "절망"
    ],
    "anger": [
        "분노", "화나", "화가 나", "짜증", "미워", "증오",
        "싫어", "원망", "폭발", "열받", "성나"
    ],
    "trust": [
        "믿음", "믿어", "신뢰", "안심", "의지", "함께", "곁에",
        "지켜줄게", "기댈", "편안"
    ],
    "anticipation": [
        "기대", "기다려", "희망", "꿈", "설렘", "두근", "바람",
        "곧", "언젠가", "다가오", "새로운 시작"
    ]
}
```

```python
def count_emotions(text: str):
    """한 곡 가사에서 각 감정별 단어 등장 횟수 세기"""
    counts = {emo: 0 for emo in emotion_lexicon.keys()}
    if not isinstance(text, str):
        return counts

    # 줄바꿈 문자 정리
    text = text.replace("\\n", " ")

    for emo, words in emotion_lexicon.items():
        for w in words:
            counts[emo] += text.count(w)
    return counts

emotion_counts = df['lyric'].apply(count_emotions)
emotion_df = pd.DataFrame(list(emotion_counts))

df_emotion = pd.concat([df.reset_index(drop=True), emotion_df], axis=1)
df_emotion.head()
```

```python
# decade별 평균 감정 점수
decade_emotions_mean = df_emotion.groupby('decade')[['joy','sadness','anger','trust','anticipation']].mean()

print(decade_emotions_mean)

plt.figure(figsize=(12,6))
for emo in ['joy','sadness','anger','trust','anticipation']:
    plt.plot(decade_emotions_mean.index, decade_emotions_mean[emo],
             marker='o', label=emo)

plt.title("K-pop 감정 스펙트럼(5감정) 10년대별 평균")
plt.xlabel("Decade")
plt.ylabel("평균 감정 단어 빈도(곡당)")
plt.grid(True)
plt.legend()
plt.show()
```
