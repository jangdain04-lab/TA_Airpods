```python
import pandas as pd

# 네 깃허브 레포의 RAW CSV 주소
url = "https://raw.githubusercontent.com/yeon524/kpop-lyrics-analytics/main/datasets/lyrics_by_year_1964_2023.csv"

# CSV 불러오기
df = pd.read_csv(url)

# 전체 가사 개수(행 수)
print("전체 가사 row 수 =", len(df))

# 컬럼 목록
print("\n컬럼 =", df.columns.tolist())

# 앞부분 미리보기
df.head()
```

```python
# 연도별 곡 수
year_counts = df['year'].value_counts().sort_index()
print(year_counts)
```
